## Now we create a Customer-Item Matrix


```python
customer_item_matrix = df.pivot_table(index='Customer_Name',columns='Seller_SKU',values='Paid_Price',aggfunc='sum')
```

```python
customer_item_matrix.head()
```

```python
customer_item_matrix = customer_item_matrix.applymap(lambda x: 1 if x>0 else 0)
customer_item_matrix.head()
```

## User based Collaborative Filtering using Sklearn module

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_item_sim_matrix = pd.DataFrame(cosine_similarity(customer_item_matrix.T))
```

```python
item_item_sim_matrix.columns = customer_item_matrix.T.index
item_item_sim_matrix['Seller_SKU'] = customer_item_matrix.T.index
item_item_sim_matrix = item_item_sim_matrix.set_index('Seller_SKU')
```

```python
item_item_sim_matrix.head()
```

```python
top_10_similar_items = list(
    item_item_sim_matrix\
            .loc['LTT12M8']\
        .sort_values(ascending=False)\
        .iloc[:10]\
    .index
)
```

```python
top_10_similar_items
```
